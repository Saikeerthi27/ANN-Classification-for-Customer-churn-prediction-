```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
import pandas as pd
import numpy as np
import pickle
```

```python
df = pd.read_csv("Churn_Modelling.csv")
```

```python
type(df)
```

```python
df.head()
```

```python
df.drop(columns = ["RowNumber",	"CustomerId",	"Surname"], axis=1, inplace=True)
```

```python
df.head()
```

```python
label_encoder_gender = LabelEncoder()
df['Gender'] = label_encoder_gender.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
ohe_encoder_geo = OneHotEncoder()
ohe_geo_encoder = ohe_encoder_geo.fit_transform(df[['Geography']])
```

```python
ohe_encoder_geo.get_feature_names_out(["Geography"])
```

```python
geo_encoded_df = pd.DataFrame(ohe_geo_encoder.toarray(), columns=ohe_encoder_geo.get_feature_names_out(["Geography"]))
```

```python
geo_encoded_df
```

```python
df.drop("Geography", axis=1, inplace=True)
```

```python
df.head()
```

```python
df = pd.concat([df, geo_encoded_df], axis=1)
```

```python
df
```

```python
# Save the encoders and scaler as pickel files

with open('label_encoder_gender.pkl', 'wb') as file:
    pickle.dump(label_encoder_gender, file) #saving

with open('ohe_encoder_geo.pkl', 'wb') as file:
    pickle.dump(ohe_encoder_geo, file)
```

```python
# Split data into indeendent and target features

X =  df.drop("EstimatedSalary", axis=1)
y = df.EstimatedSalary
```

```python
X
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=40)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
with open("scaler.pkl", 'wb') as file:
    pickle.dump(scaler, file)
```

ANN WITH REGRESSION

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
```

```python
#Step 1: Build the model
model = Sequential(
    [
        Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
        Dense(32, activation='relu'),
        Dense(1) # Here for output we dont use any activation function, we just use Linear activation function, which just returns the output
    ]
)
```

```python
# Step 2: Compile the model

model.compile(optimizer='adam', loss='mean_absolute_error', metrics=['mae'])

model.summary()
```

```python
# Step 3: Logs

from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime

log_dir = "regressionlogs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback =  TensorBoard(log_dir = log_dir, histogram_freq=1)
```

```python
# Set up early stopping

earlystopping_callback = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
# Step 4: Train the model

history = model.fit(
    X_train, y_train, validation_data=[X_test, y_test],
    epochs=100,
    callbacks=[earlystopping_callback, tensorboard_callback]

)
```

```python
# Step 5: Test the model

test_loss, test_mae = model.evaluate(X_test, y_test)
print(f'Test_mae:', {test_mae})
```
